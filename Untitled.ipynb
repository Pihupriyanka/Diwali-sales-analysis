# Gender

```python
data.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
```

```python
sales_gen =data.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
plt.show()
```

# Age

```python
sns.countplot(x='Age Group',data=data, hue ='Gender')
plt.show()
```

```python
sns.countplot(x='Age Group',data=data)
plt.show()
```

```python
state_order =data.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by = 'Orders',ascending=False).head(10)
```

```python
sns.barplot(x='State',y='Orders',data=state_order)
sns.set(rc={'figure.figsize':(20,5)})
plt.show()
```

# Marital Status

```python
sns.countplot(x='Marital_Status',data=data)
sns.set(rc={'figure.figsize':(5,10)})
plt.show()
```

```python
marital_status_order =data.groupby(['Marital_Status','Gender'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending=False).head(10)
```

```python
sns.barplot(x='Marital_Status',y='Amount',data= marital_status_order , hue='Gender')
sns.set(rc={'figure.figsize':(4,2)})
plt.show()
```

* from above graph we can conclude that married women shop the most 

# Occupation 

```python
sns.countplot(x='Occupation',data= data)
ax1 = sns.set(rc={'figure.figsize':(33,10)})
plt.show()
```

```python
occupation_orders =data.groupby(['Occupation'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending=False).head(5)
sns.barplot(x='Occupation',y='Amount',data= occupation_orders )
sns.set(rc={'figure.figsize':(10,2)})
plt.show()          
                                                                                                                                                
```

# Product_Category

```python
Product_category_orders =data.groupby(['Product_Category'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending=False).head(5)
sns.barplot(x='Product_Category',y='Amount',data= Product_category_orders )
sns.set(rc={'figure.figsize':(10,2)})
plt.show() 
```
